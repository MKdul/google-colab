```python
import pandas as pd
import numpy as np

df_1 = pd.read_csv('tmdb_5000_credits.csv')
df_2 = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_1.columns = ['id', 'title2', 'cast', 'crew']
df_2 = df_2.merge(df_1, on='id')
df_2.head(5)
```

```python
import statistics

C = df_2['vote_average'].mean()
print(C)
```

```python
m = df_2['vote_count'].quantile(0.9)
print(m)
```

```python
eligible_movies = df_2.copy().loc[df_2['vote_count']>=m]
print(eligible_movies.head)
```

```python
def weighted_rating(x, m=m, C=C):
  V = x['vote_count']
  R = x['vote_average']
  return(V/(V+m)*R)+(m/(m+V)*C)

eligible_movies['score'] = eligible_movies.apply(weighted_rating, axis=1)
```

```python
eligible_movies = eligible_movies.sort_values('score', ascending=False)
eligible_movies[['title', 'vote_count', 'vote_average', 'score']].head(20)
```

```python
import plotly.express as px

fig = px.bar((eligible_movies.head(10).sort_values('score', ascending=True)), x="score", y="title", orientation='h')
fig.show()
```

```python
df_2[['title', 'cast', 'crew', 'keywords', 'genres']].head(3)
```

```python
from ast import literal_eval

features = ['cast', 'crew', 'genres', 'keywords']

for feature in features:
  df_2[feature] = df_2[feature].apply(literal_eval)

df_2.dtypes
```

```python
def get_director(x):
  for i in x:
    if i['job'] == 'Director':
      return i['name']

  return np.nan

df_2['director'] = df_2['crew'].apply(get_director)
```

```python
def check_list(x):
  if isinstance(x, list):
    names = [i['name'] for i in x]
    return names

  return[]

features = ['cast', 'keywords', 'genres', 'crew']
for feature in features:
  df_2[feature] = df_2[feature].apply(check_list)
```

```python
df_2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
  if isinstance(x, list):
    return [str.lower(i.replace(" ", ""))for i in x]
  else:
    if isinstance(x, str):
      return str.lower(x.replace(" ", ""))
    else:
      return ""

features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
  df_2[feature] = df_2[feature].apply(clean_data)
```

```python
def create_soup(x):
  return " ".join(x['keywords'])+ ' ' + " ".join(x['cast'])+ ' ' + x['director']+ ' ' + " ".join(x['genres'])

df_2['soup'] = df_2.apply(create_soup, axis=1)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(df_2.head(2))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df_2['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df_2 = df_2.reset_index()
indices = pd.Series(df_2.index, index=df_2['title'])
```

```python
def get_recommendations(title, cosine_sim):
  idx = indices[title]
  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)
  sim_scores = sim_scores[1:11]
  movie_indices = [i[0]for i in sim_scores]
  return df_2['title'].iloc[movie_indices]
```

```python
get_recommendations("The Empire Strikes Back", cosine_sim2)
```
